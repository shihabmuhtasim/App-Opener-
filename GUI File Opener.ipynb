```python
import tkinter as tk
from tkinter import filedialog, Text
import os

#our board
root=tk.Tk() 

applist=[]

#2checking if we have the text file
if os.path.isfile('saveappnames.txt'):
    #reading the file
    with open('saveappnames.txt','r') as f:
        tempapps=f.read()   #whatever is in the test file will be here such as "...blah....,blah,"
        #making a list 
        tempapps=tempapps.split(",") 
        #striping empty spaces
        applist=[x for x in tempapps if x.strip()] 


#selecting files
def selectfile():
    #older selections destroy to avoid repeatation 
    for widget in frame_2.winfo_children():
        widget.destroy() 
    #selecting file    
    filename=filedialog.askopenfilename(initialdir='/',title="select file",filetypes=(('executables',"*.exe"),('all files','*.*')))
    applist.append(filename)
    
    #initial selected apps entry to frame 
    for app in applist:
        label=tk.Label(frame_2,text=app, bg='gray')
        label.pack()
        
#app run        
def runapp():
    for app in applist:
        os.startfile(app)
        
#add canvas 
canvas=tk.Canvas(root, height=400,width=500, bg='skyblue')#root=where we want to attach the canvas
canvas.pack() #attach

#adding a widget
frame=tk.Frame(root,bg='lightgreen') #root=where we want to attach the frame
frame.place(relwidth=0.8,relheight=0.3, relx=0.1, rely=0.1) #attach 

#frame 2 for file names
frame_2=tk.Frame(root,bg='white') #root=where we want to attach the frame
frame_2.place(relwidth=0.8,relheight=0.3, relx=0.1, rely=0.5) #attach 

#add button on root
start_app=tk.Button(root,text="Select file",padx=10,pady=5,fg='white',bg='black',command= selectfile)
start_app.pack()

#add button on frame 
run_app=tk.Button(frame,text="Run file",padx=100,pady=20,fg='white',bg='black',command= runapp)
run_app.pack()

#3entry from the text file
for app in applist:
        label=tk.Label(frame_2,text=app)
        label.pack()

root.mainloop() #our board end

#1saving app names in text file
with open('saveappnames.txt','w') as f:
    for app in applist:
        f.write(app+',')
```
